```python
def plotCorrelationMatrix(df, graphWidth):
    filename = df.dataframeName
    df = df.dropna('columns') 
    df = df[[col for col in df if df[col].nunique() > 1]] 
    if df.shape[1] < 2:
        print(f'No correlation plots shown: The number of non-NaN or constant columns ({df.shape[1]}) is less than 2')
        return
    corr = df.corr()
    plt.figure(num=None, figsize=(graphWidth, graphWidth), dpi=80, facecolor='w', edgecolor='k')
    corrMat = plt.matshow(corr, fignum = 1)
    plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
    plt.yticks(range(len(corr.columns)), corr.columns)
    plt.gca().xaxis.tick_bottom()
    plt.colorbar(corrMat)
    plt.title(f'Correlation Matrix for {filename}', fontsize=15)
    plt.show()
```

```python
plotCorrelationMatrix(df, 8)
```

## Переходим к 3 главе и построению дерева

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
X = df.drop('degradation', axis=1)
# мы выкинули из набора данных целевой признак, оставив все остальное - матрицу Х
```

```python
y = df['degradation']
```

```python
X.shape, y.shape
```

```python
# разбиваем набор данных на две подвыборки: 70/30 
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size=0.3, random_state=17) 
# установки для разбиения набора данных
```

```python
X_train.shape, X_valid.shape # смотим, какие размеры у тренировочной и тестовой(?) части
```

```python
# создаем экземпляр дерева через конструктор
first_tree = DecisionTreeClassifier(random_state=17)
```

```python
cross_val_score(first_tree, X_train, y_train, cv=5)
# указываем дерево, матрицу и вектор ответов для обучения, 5-кратная кроссвалидация
# в итоге дерево 5 раз обучилось на тестовом наборе и с такими вероятностями (0,9) получает верные ответы на оставшемся наборе данных
```

```python
np.mean(cross_val_score(first_tree, X_train, y_train, cv=5)) # среднее качество
# однако здесь дерево построено до макс. глубины, что неэффективно и может не работать на реальных данных
```

### Метод ближайших соседей

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import GridSearchCV 
# поиск по решетке, cv - кроссвалидация, грид - решетка, ищем хорошие гиперпараметры
```

```python
first_knn = KNeighborsClassifier() # конструктор создания
```

```python
# посчитаем среднюю долю верных ответов на кроссвалидации для КНН - метода ближайших соседей
np.mean(cross_val_score(first_knn, X_train, y_train, cv=5))
```

```python
# настроим гиперпараметры
knn_params = {'n_neighbors': [1, 2, 3, 4] + list(range(50, 100, 10))} 
# смотрим на соседей: от 1 до 4 и от 50 до 100 с шагом в 10
```

```python
knn_grid = GridSearchCV(first_knn, knn_params, cv=5)
```

```python
%%time
knn_grid.fit(X_train, y_train);
```

```python
knn_grid.best_score_, knn_grid.best_params_ # смотрим на лучшее качество и лучшее сочетание гиперпараметров 
```

### Дерево решений

```python
# подбор макс. глубины
tree_params = {'max_depth': np.arange(1, 11), 'max_features': np.arange(.7, 1)}
# словарь параметров дерева для перебора. мы перебираем глубину и выбираем диапазон, в данном случае он мб большим - от 1 до 10
# также перебираем макс. число признаков, которые стоит перебирать
```

```python
# создаем экземпляр для перебора по решетке для дерева
# передаем дерево, параметры, кроссвалидация=5, распараллелим выполнение (все ядра задействованы)
tree_grid = GridSearchCV(first_tree, tree_params, cv=5, n_jobs=-1)
```

```python
# передаем данные (выборку для тренировки модели) и замеряем время выполнения
#%%time
tree_grid.fit(X_train, y_train)

# получим обученный объект tree_grid, прошедший кроссвалидацию
```

```python
tree_grid.best_score_ # полученное лучшее качество на кроссвалидации,
```

```python
tree_grid.best_params_
```

```python
# посчитаем оценку на отложенной выборке
tree_grid.best_estimator_ # смотрим лучшее дерево, оно обучено на всей выборке и готово делать прогноз для отложенной
```

```python
tree_valid_pred = tree_grid.predict(X_valid) # записываем прогноз для отложенной выборки
```

```python
# надо узнать качество прогнозирования

tree_grid.score(X_valid, y_valid)
```

```python
# но лучше явно импортировать метрику - долю верных ответов
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_valid, tree_valid_pred) # доля верных ответов между реальными метками на отложенной выборке и прогнозами
```

```python
# посмотрим на долю "хороших" лезвий в отложенной выборке
1 - np.mean(y)
```

```python
# получается, что если модель прогнозирует, что лезвие всегда хорошее, то доля верных ответов - 50%
# а у нас доля верных ответов больше (99%)
```

```python
# для отрисовки деревьев импортируем метод
from sklearn.tree import export_graphviz
```

```python
export_graphviz(tree_grid.best_estimator_, out_file = 'blade_tree.dot', feature_names=X.columns, filled=True)
# передали лучшее дерево, сохранили в файл, передали названия столбцов и раскрасили вершины
```
